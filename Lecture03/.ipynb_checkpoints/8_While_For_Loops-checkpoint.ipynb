# While Statements/Loops
[Return to Table of Contents](#Table-of-Contents)

The while statement is used for repeated execution of a block of code as long as an expression is True. Usually used when the number of iterators is unknown. While loops are easier to understand, however, users need to excercise caution due to the possibility to create an infinite loop. Because of this for statements are prefered.

Documentation Reference:
- https://docs.python.org/3/reference/compound_stmts.html#the-while-statement

```python
# Initializing counter as 0. It can be defined as any number as needed.
counter = 0

# This while statement will add to the counter until the condition is met.
while counter < 5:
    counter = counter + 1
    
print(counter)
```

```python
# Counter can be initialized by any number
counter = -1

# And the addition of the counter does not need to happen in increments of 1.
while counter < 9:
    counter += 3
    
print(counter)
```

```python
# In some cases the code within the loop never runs because the condition is already met.
counter = 100

# And the addition of the counter does not need to happen in increments of 1.
while counter < 9:
    counter += 3
    
print(counter)
```

```python
# Infinite loops happen when the While statement was not develop correctly.
# You will need to use the interrupt kernel (i.e., square above next to run). 
counter = 10

# This while statement will add to the counter until the condition is met.
while counter > 5:
    counter = counter + 1
```

```python
print(f"Loop interrupted at counter = {counter}")
```

```python
# Another example of infinite loop. This time without counter.
variable_1 = 0
while variable_1 < 5 :
    print(f"Variable is {variable_1}!")
```

```python
# The loop continues until it encounters a break in one of the conditional statements.
# This example uses an If Statment inside the while loop.

passcode = "UMBC"
counter = 1

while True:
    input_word = input(f"You have three tries to enter the correct passcode: Try {counter} of 3.")
    counter = counter + 1
    if input_word == passcode:
        print("Passcode is correct!")
        break
    if input_word != passcode and counter > 3: 
        print("Sorry, Account Blocked for 100 years!")
        break
```

```python
# The same example above done a little different.
# In this example the conditional statement is defined within the while statement.
# The message is determined based on the variable trial_number rather than within the loop.

passcode = ''
trial_number = 0

while passcode != 'UMBC' and trial_number < 4:
    print(f'Enter the correct passcode in three tries: (Try {trial_number} of 3)')
    passcode = input()
    trial_number = trial_number+1
if trial_number == 4:
    print("Sorry, Account Blocked for 100 years!")
else:
    print('Passcode is correct!')
```

# While Statements/Loops: List like Data
[Return to Table of Contents](#Table-of-Contents)

In data science many tasks involve having loops to iterate thru various types of data sets. In the base Python, the most used data collections are lists and dictionaries. However, later we will explore other types of data collections and classes (each with their one advantages). Examples include:
- Arrays: list of numerical data in the Numpy library.
- Series: one dimensional array with axis labels in the Pandas library.
- Dataframes: data in table format where data is organized in columns and rows and used in the Pandas library.

The concepts and exampels in this section apply will be applicable to both base Python and other libraries data collections and classes.

```python
# Lets define a list of names.
names_list = ["Max", "Molly", "Buster", "Maggie", "Lucy", "Sam", "David"]
```

```python
# Recall that calling elements from the list as follows.
names_list[3]
```

```python
# Recall that the len() function gives us the number of elements in the list.
len(names_list)
```

```python
# The following loop iterates thru the list of names and prints them.
counter = 0

while counter < len(names_list):
    print(names_list[counter]) # In this instance the counter represents the position of the element in the list.
    counter += 1 # Adds plus one to go to the next element.
```

```python
# Let's define a list of numbers.
numbers_list = [2, 4, 6, 8, 10, 12]
len(numbers_list)
```

```python
# If I wan to create a new list of numbers where the new list is all the numbers above each increased by 2.
new_numbers_list = []
counter = 0

while counter < len(numbers_list):
    new_number = numbers_list[counter] + 2 # The new number is the number in the list plus 2.
    new_numbers_list.append(new_number)
    counter += 1
    
new_numbers_list
```

```python
# If I want to create a new list of numbers where the new list is all the numbers above and each increased by 2.
new_numbers_list = []
counter = 0

while counter <= 10: # Integer 10 goes beyond the index of the list (i.e., 6) and results in an indexerror.
    new_number = numbers_list[counter] + 2 # The new number is the number in the list plus 2.
    new_numbers_list.append(new_number)
    counter += 1
    
new_numbers_list
```

# For Statements/Loops
[Return to Table of Contents](#Table-of-Contents)

Similar to the while loop, the for statement is used for repeated execution of a block of code. However, in the for statement the loop is used to iterate over the elements of a sequence (e.g., string, list or tuple) or other iterable object. In this case the number of iterators is KNOWN and no counter is needed. The for statements has the advantage that there is no risk of having an infinite loop. For statements can also use list comprehension format.

In data science many tasks involve having loops to iterate thru various types of data sets. In the base Python, the most used data collections are lists and dictionaries. However, later we will explore other types of data collections and classes (each with their one advantages). Examples include:
- Arrays: list of numberical data in the Numpy library.
- Series: one dimensional array with axis labels in the Pandas library.
- Dataframes: data in table format where data is organized in columns and rows and used in the Pandas library.

The concepts and exampels in this section apply will be applicable to both base Python and other libraries data collections and classes. For loops are prefered because in the majority of the cases the list like data will have a KNOWN number of iterators.

Documentation References:
- https://docs.python.org/3/reference/compound_stmts.html#the-for-statement
- https://docs.python.org/3/tutorial/controlflow.html#pass-statements
- https://docs.python.org/3/library/stdtypes.html?highlight=range#range

```python
# Simple example.
for i in (1,2,3,4,5,6):
    print(i)
```

```python
# The pass statement allows to pass the loop without running it.
# This may be useful in cases where you want a placeholder and avoid empty error or indentation error.
# The pass statement can be used in while, for, and if statements.
for i in (1,2,3,4,5,6):
    pass # Future code description 
```

```python
# Indentation error when loop is empty.
for i in (1,2,3,4,5,6):
    
```

```python
# For loops in many cases can take advantage of functions such as the range function.
# https://docs.python.org/3/library/stdtypes.html?highlight=range#range
# The range constructor accepts integers.
# The first parameter is the start value (0 is default)
# The second parameter is the stop value (If 0 ValueError is raised)
# The third parameter is the step value (1 is default)

for i in range(2, 7, 1):
    print(i)
```

```python
# For loops also work with strings.
# Lets define a list of names.
names_list = ["Max", "Molly", "Buster", "Maggie", "Lucy", "Sam", "David"]

for i in names_list:
    print(i)
```

```python
a = 0
for i in range(5, 10, 1):
    a = a + i
    print(a)
```

```python
# let's add odd numbers from 1 to 9
y = 0
for x in range(1, 10, 2):
    y = y + x
    print("x is", x, " and current y is", y)
```

```python
# Calculate the number of characters in each name in the list.
for dogname in ["Max", "Molly", "Buster", "Maggie", "Lucy"]:
    print("The name", dogname,"has", len(dogname), "characters")    

print("All done.")
```

```python
names = ("Dale","Albert","Gordon","Tamara","Philip","Chester","Windom")
# In the case of names[start_element:end_element:step]
for name in names[0:5:2]:
    print(name)
```

```python
# Tv_shows, note this is a list within list.
tv_shows = [['How I Met Your Mother', 'Friends', 'Silicon Valley'],
            ['Family Guy', 'South Park', 'Rick and Morty'],
            ['Breaking Bad', 'Game of Thrones', 'The Wire']]
```

```python
# Because this is a list within list we need a nested for loop.
for sublist in tv_shows:
    for show_name in sublist:
        char_num = len(show_name)
        print("The title " + show_name + " is " + str(char_num) + " characters long.")
```

```python
# Another example of a nested loop.
for x in range(1, 4):
    for y in range(5, 9):
        print('%d * %d = %d' % (x, y, x*y))
```

```python
# Example table like data as a list within lists.
# Lists within list represents rows and elements within inner lists represents column data.
data = [[0,1,2],
        [3,4,5]]
data
```

```python
print('Number of rows=', str(len(data)))
print('Number of columns=', str(len(data[0])))
print('All the values in order first columns then rows is as follows:')

for row in data:
    for item in row:
        print(item)
```

# List Comprehension
[Return to Table of Contents](#Table-of-Contents)

Two common operations on an iterator’s output are 
1. Performing some operation for every element,
2. Selecting a subset of elements that meet some condition. 

List comprehension allow to use a loop with one line in the code for loops that are simple and only iterate thru list of elements performing some action.

Documentation Reference:
- https://docs.python.org/3/howto/functional.html?highlight=list%20comprehension#generator-expressions-and-list-comprehensions

For example, given a list of names (e.g., string), you might want to strip off trailing whitespace from each element or extract only the strings containing a given substring. In the example below a list of names have extra spaces that we want to remove. We can use the list comprehension and the strip function.

```python
names_list = ["  Max    ", "Molly    ", "     Buster", "Maggie   ", "Lucy    ", "    Sam", " David "]
```

```python
stripped_names_list = []

for element in names_list:
    stripped_names_list.append(element.strip())
                                
stripped_names_list
```

```python
# List comprehension looks like this.
stripped_names_list_comprehension = [element.strip() for element in names_list]
stripped_names_list_comprehension
```

# Other Functions
[Return to Table of Contents](#Table-of-Contents)

Python has some functions that iterate thru a data collection. There are other functions and statements that can be used with loops:
- zip() function iterates over several iterables in parallel, producing tuples with an item from each one.
- iter() function creates an iterator object.
- next() function returns the next item from the iterator. If next() is used as in the form of next(iterator, default) then "default" value is returned if the iterator is exhausted (no items left)

Doucmentation Reference:
- https://docs.python.org/3/library/functions.html#zip
- https://docs.python.org/3/library/functions.html?highlight=next#iter
- https://docs.python.org/3/library/functions.html?highlight=next#next

```python
a = [1,2,3,4,5]
b = [10,20,30,40,50]
z = zip(a,b) # Iterate over several iterables in parallel, producing tuples with an item from each one.

print(type(z))
print(z) # Sometimes the outputs are a "lazy" which means that the outputs elements won’t be processed 
# until the iterable is iterated (e.g., for loop or wrapping in a list.
print(list(z))
```

```python
# Let's explore iterators and next function.
a = list(range(3))
print(a)
```

```python
# let's create an iterator
it = iter(a)
print(it) #  Returns iterator object.
```

```python
# The next() function returns the next item from the iterator.
# if next() is used as in the form of next(iterator, default)
# then "default" value is returned if the iterator is exhausted (no items left)

# Below what we expect to print 0, 1, 2 first. Then there will be no
# more item in the iterator, so it will print "missing"
for i in range(5): # Note this range has 5 elements vs the 3 elements of list a used for the it variable.
    print(next(it, "missing"))
```

# Try Statement
[Return to Table of Contents](#Table-of-Contents)

The Try statement is used to handle errors and catch exceptions. It is very important to understand the reason of an error or exception before using the try statement to avoid the exception.

Documentation References:
- https://docs.python.org/3/reference/compound_stmts.html?highlight=try%20statement#the-try-statement
- https://docs.python.org/3/tutorial/errors.html

```python
# Recall the previous exercise were we had a list of numbers and we wante to increase each by 2.
numbers_list = [2, 4, 6, 8, 10, 12]
len(numbers_list)
```

```python
# Previously we saw the example above where we wanted to increase numbers in the list by 2.
# When we included a counter that was higher than the number of elements in list would result in an IndexError.
# Using the Try/Except statements allows us to break in the event of an indexerror.
# Note that It is very important to understand the reason of an error or exception before using the try statement.
# Not fully understanding the error/exception and using the try statement may reult in unintended outputs. 

new_numbers_list = []
counter = 0

while counter <= 10: 
    try:
        new_number = numbers_list[counter] + 2 # The new number is the number in the list plus 2.
        new_numbers_list.append(new_number)
        counter += 1
    except IndexError:
        break
    
new_numbers_list
```

# Example: Guessing Game 1: While Loop
[Return to Table of Contents](#Table-of-Contents)

```python
# Let's import Random module and the function randint():creates a random number within given parameters.
import random 
```

```python
# Guessing game
number = random.randint(1, 10)
number_of_guesses = 0

while number_of_guesses < 5:
    print('Guess a number between 1 and 10:')
    guess = input()
    guess = int(guess) # Note that I can combine this with the variable above to be more efficient.
    number_of_guesses = number_of_guesses + 1
    if guess == number:
        print('Congratulations')
        break
    if number_of_guesses == 5:
        print('Sorry, you did not guessed the number.')
        print(f'The number I was thinking is: {number}.')
```

# Example: Guessing Game 3: For Loop
[Return to Table of Contents](#Table-of-Contents)

The Age guessing game, we'll write a Python script that finds out your age in a maximum of 8 tries! Your age has to be between 0 and 128. The script asks you only one type of question: guessing your age! (e.g. “Are you 64 years old?”) and you can answer only one of these three options:
- less
- more
- correct

Based on your answer the computer can come up with another guess until it finds out your exact age.

```python
down = 0
up = 129

for i in range(1,10):
    guessed_age = int((up + down) / 2)
    answer = input('Are you ' + str(guessed_age) + "years old? (Type 'less', 'more' or 'correct'")
    if answer.lower() == 'correct':
        print("Nice")
        break
    elif answer.lower() == 'less':
        up = guessed_age
    elif answer.lower() == 'more':
        down = guessed_age
    else:
        print('Wrong answer')
```
