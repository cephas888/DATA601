#### Building a widget
[Table of Contents](#Table-of-Contents)

There are multiple ways to build and run widgets. Two main elements is the interact_manual and the interact functions. The  "interact_manual" creates a "Run Interact" button and does not run the specified functions until the button is selected. The  "interact" updates the function as we make changes in the widget selection. There are other elements that may be optional such as defining a variable name or the way that the functions are built and called.  

In this example we will create a function that takes in an input text and prints a line with some other text and the input text. One exmaple with the interact_manual and the same example with the interact. Note that for functions that take some time to run you want to use the "interact_manual" as making small changes in the "interact" make result in freezing or having to wait the processor to solve the functions.

In general building a widget follows this steps.
1. Define funcitons you will be using or want to run (e.g., filter data, calculation, etc) with the input parameters that you will feed into the function.
2. Use either interact_manual or interact function to define the app variable
    a. Call the defined function in the first parameter
    b. Define which widget the function will use as inputs.
    c. Define parameters for each widget.
    
Documentation References:
- [Widget Basics](https://ipywidgets.readthedocs.io/en/7.7.0/examples/Widget%20Basics.html)
- [Using Interact](https://ipywidgets.readthedocs.io/en/7.6.2/examples/Using%20Interact.html)

```python
def print_data(text_to_print):
    print(f'The text that you want to print is: "{text_to_print}"')
```

```python
# The interact manual does not call the function and applies the input parameters until "Run Interact" button is selected.
interact_manual(print_data,
                text_to_print = widgets.Text(value='',
                                             placeholder='Type Something',
                                             description='Input Text Box:',
                                             disabled=False,
                                             style={'description_width': 'initial'}, 
                                             layout = widgets.Layout(width='50%')));
```

```python
# We can also define the interact_manual or interact as a variable.
print_app = interact_manual(print_data,
                            text_to_print = widgets.Text(value='',
                                                         placeholder='Type Something',
                                                         description='Input Text Box:',
                                                         disabled=False,
                                                         style={'description_width': 'initial'}, 
                                                         layout = widgets.Layout(width='50%')));
```

```python
# Note that the interact will show the changes interactively.
print_app_dynamic = interact(print_data,
                             text_to_print = widgets.Text(value='',
                                                 placeholder='Type Something',
                                                 description='Input Text Box:',
                                                 disabled=False,
                                                 style={'description_width': 'initial'}, 
                                                 layout = widgets.Layout(width='50%')))
```

#### Unique values of dropdown menus
[Table of Contents](#Table-of-Contents)

```python
# List of the Release Calendar Years
release_cy_list = (df_data['release_cy'].unique()).tolist() # Extract unique calendar year values.
release_cy_list.sort() # Sorts the release calendar year values
#release_cy_list = ['NaT']+release_cy_list # Includes 'NaT'. Only needed if data has undefined year.
release_cy_list = list(dict.fromkeys([element for element in release_cy_list])) # Removes and Duplicates
#print(release_cy_list)
```

```python
# Develops a unique list of genres to be able to use it in the menus.
# Note that this is using the one-hot-encoded Genres column features to extract the options.

# First and Last index of the Movie Genres. 
# Two methods: First using the columns of the main dataframe, the other of the dfcolumns_index dataframe.
first_genre_index = df_data.columns.to_list().index("Action")
#first_genre_index = dfcolumns_index.index[dfcolumns_index['column'] == str('Action')][0] 

last_genre_index = df_data.columns.to_list().index("Western")
#last_genre_index = dfcolumns_index.index[dfcolumns_index['column'] == str('Western')][0] 

# List of Unique Genres.
genres_list = sorted(dfcolumns_index['column'][first_genre_index:last_genre_index+1].to_list())
#genres_list
```

#### Filter and Plot Function
[Table of Contents](#Table-of-Contents)

```python
def filter_plot_data(vote_count_filter, box_office_earning_filter, release_cy_filter, genres_filter):
    
    # These lines of code are used to show how each Widget outputs the data.
    display(HTML(f'<h2>These lines of code are used to show how each Widget outputs the data:<h2>'))
    print(f'1. vote_count_filter is submitted as a {type(vote_count_filter)} data type. Selection is: {vote_count_filter}.')
    print(f'2. box_office_earning_filter is submitted as a {type(box_office_earning_filter)} data type. Selection is:  {box_office_earning_filter}.')
    print(f'3. release_cy_filter is submitted as a {type(release_cy_filter)} data type. Selection is: {release_cy_filter}.')
    print(f'4. genres_list is submitted as a {type(genres_filter)} data type. Selection is: {genres_filter}')
    
    display(HTML(f'<h2>NOTES:<h2>'))
    print('When selecting all filters the source data and filtered data shapes do not match.')
    print('This movie can be found in the data science example dashboard by using the NOT and selecting all genres.')
    print('This movie is an example of getting to the dashboard creation and finding an item that could potentially be cleaned out in the data cleaning notebook.')

        # Note that there is one movie that has the Genre's missing and needs to be cleaned out.
    # This movie is an example of getting to the dashboard creating and finding this item to be cleaned.
    
    # Filterieng of the data using the functions
    df_data_filtered = df_data.loc[(df_data['vote_count'] >= min(vote_count_filter)) &
                                   (df_data['vote_count'] <= max(vote_count_filter)) &
                                   (df_data['box_office_earning'] >= box_office_earning_filter) &
                                   (df_data['release_cy'] >= min(release_cy_filter)) &
                                   (df_data['release_cy'] <= max(release_cy_filter)) &
                                   (df_data['genres'].str.contains('|'.join(genres_filter), 
                                                                   flags=re.IGNORECASE, regex=True))
                                  ].copy().reset_index(drop = True)
    
    display(HTML(f'<h2>Selected Data Rows:<h2>'))
    print(f'The fitlered data contains {df_data_filtered.shape[0]} rows of data.')
    print(f'The source data has {len(df_data)} records and ONLY {len(df_data_filtered)} records are selected.')

    display(HTML(f'<h2>Scatter Plot box_office_earning vs. runtime<h2>'))
    plt.scatter(x = df_data_filtered['box_office_earning'], 
                y = df_data_filtered['runtime'], 
                alpha = 0.25)
    plt.show()
```
